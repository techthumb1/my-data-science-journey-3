# Dimensionality Reduction: PCA, t-SNE, and LDA
This notebook demonstrates dimensionality reduction techniques on a labeled dataset:
- PCA: Principal Component Analysis
- t-SNE: t-Distributed Stochastic Neighbor Embedding
- LDA: Linear Discriminant Analysis

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.preprocessing import StandardScaler
```

```python
# Load and standardize the digits dataset
digits = load_digits()
X = digits.data
y = digits.target
X_scaled = StandardScaler().fit_transform(X)
```

## PCA

```python
# Apply PCA and plot explained variance
pca = PCA(n_components=10)
X_pca = pca.fit_transform(X_scaled)
plt.figure(figsize=(8, 5))
plt.bar(range(1, 11), pca.explained_variance_ratio_, alpha=0.7, align='center')
plt.ylabel('Explained Variance Ratio')
plt.xlabel('Principal Components')
plt.title('PCA Explained Variance')
plt.show()
```

```python
# PCA 2D visualization
plt.figure(figsize=(8, 5))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='tab10', s=15)
plt.title('PCA Projection (2D)')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```

## t-SNE

```python
# Apply t-SNE
tsne = TSNE(n_components=2, random_state=42, perplexity=30)
X_tsne = tsne.fit_transform(X_scaled)
plt.figure(figsize=(8, 5))
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap='tab10', s=15)
plt.title('t-SNE Projection')
plt.xlabel('t-SNE 1')
plt.ylabel('t-SNE 2')
plt.show()
```

## LDA

```python
# Apply LDA
lda = LDA(n_components=2)
X_lda = lda.fit_transform(X_scaled, y)
plt.figure(figsize=(8, 5))
plt.scatter(X_lda[:, 0], X_lda[:, 1], c=y, cmap='tab10', s=15)
plt.title('LDA Projection (2D)')
plt.xlabel('LD1')
plt.ylabel('LD2')
plt.show()
```
